## 1. Imports de base

Librairies Python standard pour ce cas d'usage.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
np.random.seed(42)
print(f'pandas {pd.__version__} | numpy {np.__version__}')
```

## 2. Donnees simulees

Generation d'un dataset de demonstration. Toutes les donnees sont simulees.


```python
N = 1000
np.random.seed(42)
X = pd.DataFrame(np.random.randn(N, 5),
                  columns=['feature_1','feature_2','feature_3','feature_4','feature_5'])
X['target'] = (X['feature_1'] - X['feature_2'] + np.random.normal(0, 0.3, N) > 0).astype(int)
print(f'Dataset: {N} observations')
print(X.describe().round(3))
```

## 3. Exploration rapide

Statistiques descriptives et visualisation.


```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
X['feature_1'].hist(bins=40, ax=axes[0], color='steelblue', alpha=0.7)
axes[0].set_title('Distribution feature_1')
X.corr().iloc[:5,:5].style.background_gradient(cmap='coolwarm')
print('Correlations:')
print(X.iloc[:,:5].corr().round(3))
plt.tight_layout()
plt.show()
print('Exploration complete')
```
